#### Line Graph

```python
# Plotting Line Graph With Matplotlib
%matplotlib inline

# importing the required module 
import matplotlib.pyplot as plt 
  
# x axis values 
x = [1,2,3] 
# corresponding y axis values 
y = [10,9,8] 
  
# plotting points  
plt.plot(x, y) 
  
# name x axis 
plt.xlabel('x - axis') 
# name y axis 
plt.ylabel('y - axis') 
  
# title to my graph 
plt.title('Line Graph!') 
  
# show the plot 
plt.show() 
```

```python
# Plotting Multiple Lines On A graph

x1 = [1,2,3]
y1 = [10,9,8]

x2 = [1,2,3]
y2 = [8,9,10]

# plotting points  
plt.plot(x1, y1,label= 'Line 1')
plt.plot(x2, y2,label='Line 2')
  
# name x axis 
plt.xlabel('x - axis') 
# name y axis 
plt.ylabel('y - axis') 
  
# title to my graph 
plt.title('Multiple Line Graph!') 

plt.legend()
  
# show the plot 
plt.show() 
```

```python
%matplotlib inline
caps=['Hyderabad',
 'Patna',
 'Agartala',
 'Kohima',
 'Chandigarh',
 'Ranchi',
 'Thiruvananthapuram',
 'Mumbai',
 'Bhubaneswar',
 'Chandigarh',
 'Panaji',
 'Gandhinagar',
 'Dispur',
 'Amaravati',
 'Dehradun',
 'Shimla',
 'Bhopal',
 'Jaipur',
 'Aizawl',
 'Srinagar',
 'Kolkata',
 'Raipur',
 'Chennai',
 'Lucknow',
 'Gangtok',
 'Itanagar',
 'Bangalore',
 'Shillong',
 'Imphal']
import requests
import json
temp = []
city = []
weather = []
k = 0
for i in caps:
    z=requests.get('http://api.openweathermap.org/data/2.5/weather?q='+i+'&appid=770297e3792414ec205e9282b0ca25cb')
    dic=json.loads(z.content.decode())
    if dic['cod'] == 200:
        k+=1
        print('City Name:',dic['name'])
        print('Temperature:',dic['main']['temp'])
        print('Weather:',dic['weather'][0]['main'])
        temp.append(dic['main']['temp'])
        city.append(dic['name'])
        weather.append(dic['weather'][0]['main'])
        
    else:
        print(i,'Spelling is Incorrect')
import matplotlib.pyplot as plt
plt.figure(figsize=(10,10))
plt.plot(temp,city,color='green', linestyle='dashed', linewidth = 3, 
         marker='o', markerfacecolor='blue', markersize=10)
plt.xlabel('Temperature In Kelvin(K)')
plt.ylabel('State Capitals Of India')
plt.title('Temperature Variation By State Capitals Of IN')
plt.show()
import matplotlib.pyplot as plt
plt.figure(figsize=(10,10))
plt.barh(city,temp)
plt.xlabel('Temperature In Kelvin(K)')
plt.ylabel('State Capitals Of India')
plt.title('Temperature Variation By State Capitals Of IN')
plt.show()
import matplotlib.pyplot as plt
plt.figure(figsize=(8,8))
plt.scatter(temp,city,alpha=0.8,cmap='green',marker='*',edgecolors='green',linewidths=1.0,c='blue',s=100)
plt.xlabel('Temperature In Kelvin(K)')
plt.ylabel('State Capitals Of India')
plt.title('Temperature Variation By State Capitals Of IN')
plt.show()
```
